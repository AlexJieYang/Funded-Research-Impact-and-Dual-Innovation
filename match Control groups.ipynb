```python
def get01(x):
    if x == 0:
        return 0
    else:
        return 1
df1['NIH']  = df1['NIH_Count'].apply(get01)
df1['NSF']  = df1['NSF_Count'].apply(get01)
```

```python
for column in ['Volume', 'Issue', 'FirstPage', 'LastPage']:
    df1[column] = df1[column].fillna(0)
```

```python
df2 = df1[df1['NIH']==1].copy()
df2.index = range(len(df2))
```

```python
merged_df = df1.merge(df2, on=['Year', 'FieldID', 'JournalID', 'Volume', 'Issue'], how='inner')

# for i in df2.index():
#     Year = df2.Year.values[i]
#     FieldID = df2.FieldID.values[i]
#     JournalID = df2.JournalID.values[i]
#     Volume = df2.Volume.values[i]
#     Issue = df2.Issue.values[i]
#     df1.loc[(df1['Year']==Year) &(df1['FieldID']==FieldID) &(df1['JournalID']==JournalID) &(df1['Volume']==Volume)&(df1['Issue']==Issue) ,:]
```
